```python
import csv

import matplotlib.pyplot as plt

from collections import Counter


# NOTE:
#   - `EVAL0_NUM` is for initial evaluation (i.e., making sure IAA is high)
#   - `EVAL1_NUM` is for the final evaluation
EVAL0_NUM: int = 16
EVAL1_NUM: int = 64

with open("counts.csv", encoding="utf-8") as file:
    reader = csv.reader(file)
    header = next(reader)
    counts = Counter({name: int(freq) for name, freq in reader})

top4 = dict(counts.most_common(4))
top4
```

```python
objects = top4.keys()
y_pos = range(len(objects))
x_pos = top4.values()

plt.style.use("bmh")

plt.bar(y_pos, x_pos, align="center", alpha=0.5)
plt.xticks(y_pos, objects, fontsize="6")
plt.ylabel("Frequency")
plt.title("Rare Disease Distribution (Weak Supervision Rules)")

plt.savefig("distribution.png", dpi=256)
```
